## 1.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/airbnb_hw.csv', low_memory=False)
df.head()
```

```python
price = df['Price']
price
```

```python
price = price.replace(',','', regex=True)
price = pd.to_numeric(price, errors="coerce")
df['Price'] = price
sum(df['Price'].isnull())
```

The sum of isnull() = 0 meaning it produced all 0s therefore there are no missing values

## 2.

```python
df = pd.read_csv('./data/sharks.csv', low_memory=False)
df.head()
```

```python
typeCol = df['Type']
typeCol
```

```python
typeCol.unique()
```

```python
typeCol.value_counts()
```

```python
typeCol = typeCol.replace(['Watercraft', 'Boating', 'Sea Disaster', 'Boat', 'Boatomg'], 'Water Disaster')
typeCol = typeCol.replace(['Questionable', 'Unconfirmed', 'Under investigation', 'Unverified', 'Invalid'], np.nan)
```

```python
typeCol.value_counts()
```

```python
df['Type'] = typeCol
```

I chose to reduce the different types of shark attacks to these three categories Unprovoked, Provoked, and Water Disaster because the first two were already included and meaningful. This is because we might want to look at the correlations between a provoked or unprovoked attack and the other data. As for "Water Disaster," the types generalized to this new type were all either synonymous or misspellings of another. The other types that were removed were done so because they were labeled either as inconclusive or invalid so we marked them all with NaNs because we don't want to use any invalid data when analyzing.

## 3.

```python
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False)
df.head()
```

```python
released_pretrial = df['WhetherDefendantWasReleasedPretrial']
released_pretrial.value_counts()
```

Label 9 - Insufficient/missing information therefore we will remove these

```python
released_pretrial = released_pretrial.replace(9, np.nan)
print(released_pretrial.value_counts())
print(f"\nTotal missings: {sum(released_pretrial.isnull())}")

df['WhetherDefendantWasReleasedPretrial'] = released_pretrial
```

## 4.

```python
imposed_sentence = df['ImposedSentenceAllChargeInContactEvent']
imposed_sentence = pd.to_numeric(imposed_sentence,errors='coerce')

imposed_sentence_missing = imposed_sentence.isnull()
print(f"Missing data: {sum(imposed_sentence_missing)}")
```

```python
sentence_type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
sentence_type.value_counts()
```

```python
pd.crosstab(imposed_sentence.isnull(), sentence_type)
```

4 - "deferred, etc." which means the charges were deferred and they served no sentence which is the reason for the NaN values in imposed_sentence. We will instead replace these NaNs with 0s for each row with a sentence type of 4 to accuretly reflect the time they served and reduce the number of missing data.

```python
# df.loc[ condition, var] = value
df.loc[(sentence_type == 4), 'ImposedSentenceAllChargeInContactEvent'] = 0

imposed_sentence = df['ImposedSentenceAllChargeInContactEvent']
imposed_sentence = pd.to_numeric(imposed_sentence,errors='coerce')

pd.crosstab(imposed_sentence.isnull(), sentence_type)
```

```python
df['ImposedSentenceAllChargeInContactEvent'] = imposed_sentence
```
